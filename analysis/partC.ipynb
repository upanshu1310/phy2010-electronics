```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfs = pd.read_csv('../data/partC/series.csv')
dfp = pd.read_csv('../data/partC/parallel.csv')
```

```python
L = 0.03
C = 0.0000001
R = 100
```

```python
w0 = (1/(L*C)**0.5)/(2*np.pi)
Rmax = 2*(L/C)**0.5
```

```python
w0
np.log10(w0)
```

```python
R < Rmax
```

```python
dfs.T[0]
```

```python
plt.scatter(dfs['F_log'],dfs['Gain_dB'],s=8,label='data')
plt.xlabel('Frequency (log)')
plt.ylabel('Gain (dB)')
plt.axvline(np.log10(w0),color='red',label='resonant frequency')
plt.legend()
plt.show()
```

```python
plt.scatter(dfp['F_log'],dfp['Gain_dB'],s=8,label='data')
plt.xlabel('Frequency (log)')
plt.ylabel('Gain (dB)')
plt.axvline(np.log10(w0),color='red',label='resonant frequency')
plt.legend()
plt.show()
```

```python
plt.scatter((dfp['F_log']+dfs['F_log'])/2,dfp['Gain_dB']+dfs['Gain_dB'],s=2)
plt.xlabel('Frequency (log)')
plt.ylabel('Gain (dB)')
plt.axvline(np.log10(w0),color='red',label='cutoff frequency')
plt.legend()
plt.show()
```
