```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfi = pd.read_csv('./../data/partA/integrator.csv')
dfdt = pd.read_csv('./../data/partA/differentiator_triangle.csv')
dfds = pd.read_csv('./../data/partA/differentiator_square.csv')
```

```python
R = 10*10e3
C = 0.022/10e6

w = 1/(R*C*2*np.pi)
```

```python
plt.scatter(dfi['F_log'],dfi['Gain_dB'],s=2,label='data')
plt.xlabel('Frequency (log)')
plt.ylabel('Gain (dB)')
plt.axvline(np.log10(w),color='red',label='cutoff frequency')
plt.legend()
plt.show()
```
