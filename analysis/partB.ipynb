```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfi = pd.read_csv('./data/partB/integrator.csv')
dfd = pd.read_csv('./data/partB/differentiator.csv')
dfi.T[0]

R = 10*10e3
C = 0.022/10e6
```

```python
def cutoff(R,C):
    return 1/(R*C*2*np.pi)
```

```python
w = cutoff(R,C)
```

```python
plt.scatter(dfi['F_log'],dfi['Gain_dB'],s=8,label='data')
plt.xlabel('Frequency (log)')
plt.ylabel('Gain (dB)')
plt.axvline(np.log10(w),color='red',label='cutoff frequency')
plt.legend()
plt.show()
```

```python
plt.scatter(dfd['F_log'],dfd['Gain_dB'],s=8,label='data')
plt.xlabel('Frequency (log)')
plt.ylabel('Gain (dB)')
plt.axvline(np.log10(w),color='red',label='cutoff frequency')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(11,3))

ax[0].scatter(dfi['F_log'],dfi['Gain_dB'],s=2,label='data')
ax[0].set_xlabel('Frequency (log)')
ax[0].set_ylabel('Gain (dB)')
ax[0].axvline(np.log10(w),color='red',label='cutoff frequency')

ax[1].scatter(dfd['F_log'],dfd['Gain_dB'],s=2,label='data')
ax[1].set_xlabel('Frequency (log)')
ax[1].set_ylabel('Gain (dB)')
ax[1].axvline(np.log10(w),color='red',label='cutoff frequency')

plt.tight_layout()
plt.show()
```
